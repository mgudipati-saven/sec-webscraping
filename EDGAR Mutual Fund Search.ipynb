```python
# import our libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# let's first make a function that will make the process of building a url easy.
def make_url(base_url , comp):
    url = base_url
    
    # add each component to the base url
    for r in comp:
        url = '{}/{}'.format(url, r)
        
    return url
```

```python
# base URL for the SEC EDGAR browser
endpoint = r"https://www.sec.gov/cgi-bin/browse-edgar"

#NPORT&dateb=&count=40&scd=filings&search_text="

# define our parameters dictionary
param_dict = {
              'action':'getcompany',
              'CIK':'C000143077',
              'type':'NPORT-P',
              'owner':'exclude',
              'output':'atom',
              'count':'100'
             }

# request the url, and then parse the response.
response = requests.get(url = endpoint, params = param_dict)
soup = BeautifulSoup(response.content, 'xml')

# Let the user know it was successful.
print(response.status_code)
print(response.url)
```

```python
# find all the entry tags
entries = soup.find_all('entry')
filing_url = entries[0].find('filing-href').text
print(filing_url)
```

```python
idx = filing_url.rindex('/')
url = filing_url[0:idx] + '/primary_doc.xml'

# request the url, and then parse the response.
response = requests.get(url)
soup = BeautifulSoup(response.content, 'xml')

# Let the user know it was successful.
print(response.status_code)
print(response.url)
```

```python
holdings = soup.find_all('invstOrSec')
```

```python
for holding in holdings:
    name = holding.find('name').text
    title = holding.find('title').text
    pct = holding.find('pctVal').text
    value = holding.find('valUSD').text
    balance = holding.find('balance').text
    
    print(name, title, value, pct, balance)
```
